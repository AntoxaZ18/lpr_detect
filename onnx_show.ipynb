```python
from yolo_onnnx import YoloONNX
import cv2
from time import time
import os
from matplotlib import pyplot as plt
```

```python
model = './onnx_models/baseline_nano_.onnx'


model = YoloONNX(model, device='cpu', classes=['LPR'])
frame = cv2.imread('night.jpg')

results = model([frame])

plt.figure(figsize=(20, 12))
plt.imshow(results[0], interpolation='nearest')
plt.show()
```
